# Input training datapoints

```python
import numpy as np

X1 = np.array([[2,3,4,3,5,5,2,4,0,6], [3,2,4,2,4,3,8,6,0,5]])
X2 = np.array([[7,9,10,11,9,13,9,11,11,10], [16,15,13,11,17,13,14,15,14,12]])
Xtest = np.array([[4, 10, 10], [0, 2.5, 12.5]])
```

# Import functions from LDA.py

```python
from LDA import LDA, plot_LDA, plot_pdfs, plot_posteriors
```

# Train model

```python
mean1, mean2, S1, S2, Sw, Swinv, w, Y1, Y2, mu_Y1, mu_Y2, Var_Y1, Var_Y2, a, b, c, s1, s2 = LDA(X1, X2)
print("mean1:\n",mean1)
print("mean2:\n",mean2)
print("Scatter Matrix S1:\n", S1)
print("Scatter Matrix S2:\n", S2)
print("Within-class Scatter Matrix Sw\n", Sw)
print("Between-class Scatter Matrix Swinv\n", Swinv)
print("Optimum Projection Vector w\n", w)
print("Projected Space of Class1 Y1\n", Y1)
print("Projected Space of Class2 Y2\n", Y2)
print("mu_Y1:",mu_Y1)
print("mu_Y2:",mu_Y2)
print("Var_Y1:",Var_Y1)
print("Var_Y2:",Var_Y2)
print("Classification threshold a:",a)
print("Classification threshold b:",b)
print("Classification threshold c:",c)
print("Decision Threshold s1:",s1)
print("Decision Threshold s2:",s2)
```

# Plot training datapoints with projection vector superimposed

```python
plot_LDA(X1, X2, w)
```

# Plot f(y|Ck) for k = 1,2

```python
plot_pdfs(mu_Y1, mu_Y2, Var_Y1, Var_Y2)
```

# # Plot f(Ck/y) for k = 1,2

```python
plot_posteriors(mu_Y1, mu_Y2, Var_Y1, Var_Y2)
```

# Classify test datapoints

```python
# project test data points into the projected space
Ytest = np.dot(w.T, Xtest)

# calculate mean of class 1 and class 2 in the projected space
mean_Y1 = np.mean(Y1.reshape(-1, 1))
mean_Y2 = np.mean(Y2.reshape(-1, 1))

# initialize list to store class labels of test data points
labels = []

# loop over each test data point
for y in Ytest.T:
    # calculate distances of the projected test data point from the means
    dist1 = np.linalg.norm(y - mean_Y1)
    dist2 = np.linalg.norm(y - mean_Y2)
    
    # assign the test data point to the class whose mean is closer
    if dist1 < dist2:
        labels.append('Class 1')
    else:
        labels.append('Class 2')

print(labels)
```
